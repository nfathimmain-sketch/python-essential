Challenge 2 (10 Marks)
Customer Feedback Analyzer
Scenario

An e-commerce company records customer ratings after each purchase.

The rating is converted into feedback labels.

Rating	Feedback
5	Excellent
4	Good
3	Average
1-2	Poor

The program should process multiple customer records and display the final feedback summary.



```python
def get_feedback(rating):
    # TODO 1
    # Return feedback
    if rating == 5:
        return "Excellent"
    elif rating == 4:
        return "Good"
    elif rating == 3:
        return "Average"
    else:
        return "Poor"

def feedback_summary():
    n = int(input("Enter number of customers: "))

    feedback = {}

    for i in range(n):
        customer = input()
        rating = int(input())

        # TODO 2
        # Find feedback
    result = get_feedback(rating)

        # TODO 3
        # Store customer and feedback
    result= feedback[customer]

    print("\nCustomer Feedback Summary")

    # TODO 4
    # Display all customer feedback
    for customer in feedback:
     print(customer, ":", feedback[customer])

feedback_summary()


# Sample Input
# 3
# Alice
# 5
# John
# 3
# Maria
# 2
# Output
# Customer Feedback Summary

# Alice : Excellent
# John : Average
# Maria : Poor
```
