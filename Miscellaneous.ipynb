# To handle marks of 4 subjects for 3 students

```python
marks_students=[
    [50,30,60,45],
    [80,40,55,67],
    [55,45,65,75]
]
print(marks_students)
print(marks_students[2][2])#3rd subject marks for 3 students
for i in range(3):
  print(marks_students[i])
  for j in range(4):
    print(marks_students[i][j])
```

```python
students = [
    {"roll": 1, "name": "Fathima", "python": 80, "ds": 75, "maths": 90},
    {"roll": 2, "name": "Rahul", "python": 70, "ds": 65, "maths": 60},
    {"roll": 3, "name": "Anu", "python": 90, "ds": 85, "maths": 88}
]

for s in students:
    print(s["roll"], s["name"])
r = int(input("Enter roll number:"))

for s in students:
    if s["roll"]==r:
        print("Found:", s)
 max_total = 0
topper = None

for s in students:
    total = s["python"] + s["ds"] + s["maths"]
    if total > max_total:
        max_total = total
        topper = s

print("Topper:", topper["name"])
```
